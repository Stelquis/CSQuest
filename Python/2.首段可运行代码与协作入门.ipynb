## 第 1 节 · `print` 与输出

`print` 是 Python 中最常用的输出函数。程序与外界的第一次对话，通常从这里开始。

### 1.1 基础用法

```python
print("Hello, I'm Yibo! I love python.")          # 输出字符串
print(42)               # 输出整数
print(3.14)             # 输出浮点数
print("a", "b", "c")    # 输出多个值，默认用空格分隔
```

### 1.2 `sep` 与 `end`

| 参数 | 默认值 | 作用 |
|:---|:---|:---|
| `sep` | 空格 `' '` | 多个值之间的分隔符 |
| `end` | 换行 `'\n'` | 输出末尾字符 |

## 第 2 节 · 注释的标准写法

注释是写给**未来的自己和其他人**看的，不是写给解释器看的。

### 2.1 两种注释

```python
# 这是单行注释：说明「为什么」这么做

"""
这是多行字符串，常用作函数说明（docstring）。
初学阶段知道即可，不必强行每行都写。
"""
```

### 2.2 该写 vs 不该写

| 该写 | 不该写 |
|:---|:---|
| 解释非显而易见的业务规则 | `# 把 x 加 1`（代码已经说明了） |
| 标注已知限制或临时方案 | 大段注释掩盖烂代码 |
| 说明复杂公式的来源 | 把注释当聊天记录 |

## 第 3 节 · 如何下精确指令

工具不是读心者。你给的指令越模糊，产出越随机。

### 3.1 精确指令的五要素

| 要素 | 问题 | 示例 |
|:---|:---|:---|
| **任务目标** | 要做什么？ | 写一个函数，统计列表中正数的个数 |
| **输入** | 参数是什么类型？ | 参数 `nums: list`，元素为 int 或 float |
| **输出** | 返回什么？ | 返回 int，表示正数个数 |
| **约束** | 有什么限制？ | 不使用第三方库；空列表返回 0 |
| **验收标准** | 怎样算通过？ | `count_positive([1,-2,3])` 返回 2；`count_positive([])` 返回 0 |

### 3.2 模糊 vs 精确

| 模糊（差） | 精确（好） |
|:---|:---|
| 帮我写个处理数据的程序 | 用 Python 读取同目录下 `sales.csv`，打印前 5 行，使用 pandas |
| 做个计算器 | 写函数 `add(a, b)` 返回两数之和，参数为 int，不处理非数字输入 |
| 优化一下这段代码 | 在不改变输出的前提下，把下面函数的行数从 20 行减到 10 行以内：[贴代码] |

## 第 4 节 · 好代码与坏代码

### 4.1 评判维度

| 维度 | 好代码 | 坏代码 |
|:---|:---|:---|
| **可读性** | 变量名有意义 | `a`, `tmp`, `data1` |
| **可预测** | 同样输入同样输出 | 依赖全局变量，结果随运行顺序变 |
| **边界处理** | 考虑空输入、异常 | 默认输入永远正常 |
| **适度** | 能读懂的最简实现 | 一行写 200 字符 |

### 4.2 对照案例


````python
# 坏代码：名字无意义，无边界
def f(l):
    return len([i for i in l if i>0])

# 好代码：名字清晰，文档说明，边界明确
def count_positive(numbers):
    """返回列表中正数的个数。空列表返回 0。"""
    count = 0
    for n in numbers:
        if n > 0:
            count += 1
    return count

# 验收
print(count_positive([1, -2, 3, 0]))  # 期望 2
print(count_positive([]))             # 期望 0
````
