## 第 1 节 · 计算机如何执行程序

很多人把「写代码」理解成「跟电脑聊天」。不对。

电脑只做一件事：**按顺序执行你写好的指令**。它不会猜你的意图，也不会替你补全你没写清楚的步骤。

### 1.1 三个核心概念

| 概念 | 含义 | 类比 |
|:---|:---|:---|
| **存储** | 数据放在内存或磁盘里 | 厨房里的食材柜 |
| **指令** | 告诉计算机做什么 | 菜谱上的步骤 |
| **执行顺序** | 从上到下、按条件分支 | 先洗菜，再切菜，最后下锅 |

一段最简单的 Python 程序：

```python
name = "学员"
print("你好，" + name)
```

执行过程拆解：

1. 在内存中开辟空间，存入字符串 `"学员"`，贴上标签 `name`
2. 调用 `print`，把 `"你好，" + name` 的结果输出到屏幕
3. 程序结束

**关键点**：第二行不会自己运行，除非第一行已经执行完毕。顺序错了，结果就错。

## 第 2 节 · Python 在技术栈中的位置

Python 不是「最好的语言」，但它是**当前技术栈里覆盖面最广的胶水语言**。

### 2.1 各语言典型分工

| 语言 / 技术 | 擅长的事 | 不擅长的事 |
|:---|:---|:---|
| **C / C++** | 操作系统、游戏引擎、极致性能 | 快速写业务脚本 |
| **Java** | 大型企业后端、Android | 数据分析原型 |
| **JavaScript** | 浏览器交互、前端界面 | 离线科学计算 |
| **Python** | 脚本自动化、数据分析、AI 原型、教学 | 手机端原生 App |

### 2.2 为什么 AI 时代更要学 Python

三个硬理由：

1. **生态**：数据科学、机器学习库的第一入口仍是 Python
2. **可读性**：语法接近自然语言，适合快速验证想法
3. **协作接口**：无论用什么工具生成代码，最终落地、调试、验收，都需要人读懂 Python

学 Python 不是为了跟工具抢活，是为了**能验收工具产出的东西**。

## 第 3 节 · 生成式工具写代码的系统性短板

工具能写代码，不等于能交付代码。在工程现场，以下几类问题反复出现。

### 3.1 四类常见失败

| 类型 | 表现 | 后果 |
|:---|:---|:---|
| **幻觉 API** | 调用不存在的函数或参数 | 运行直接报错 |
| **忽略边界** | 没处理空值、零除、文件不存在 | 特定输入下崩溃 |
| **逻辑看似正确** | 能跑，但结果不对 | 最难发现，最危险 |
| **不可维护** | 变量名 `a1`、`tmp`，无注释 | 两周后自己也看不懂 |

### 3.2 真实案例

需求：「统计一个列表里正数的个数」

工具可能生成：

```python
def count_pos(lst):
    return len([x for x in lst if x > 0])
```

问题：`lst` 里如果有 `None` 或非数字，比较 `x > 0` 会抛异常。工具没问输入约束，也没写防御。

**人该补的一步**：先确认输入类型，或在验收时构造边界用例。

## 第 4 节 · 人掌握代码的核心价值

在团队协作中，代码最终要对**人**负责，不对工具负责。

### 4.1 四项不可替代的能力

| 能力 | 具体含义 |
|:---|:---|
| **规格澄清** | 把「大概想要」变成「输入是什么、输出是什么、边界是什么」 |
| **验收** | 用测试用例证明代码在边界条件下也正确 |
| **调试** | 读报错、定位行号、逐行推理状态变化 |
| **责任归属** | 上线出问题，查的是提交记录和作者，不是「工具当时怎么生成的」 |

### 4.2 一句话原则

> 工具负责提速，人负责正确。

## 第 7 节 · 三条环境路径与决策

### 7.1 路径对照

| 路径 | 适合谁 | 优点 | 缺点 |
|:---|:---|:---|:---|
| **Windows 本地** | 日常用 Windows 的学员 | 完全离线、性能好 | 安装步骤多，PATH 易出问题 |
| **Mac 本地** | 日常用 Mac 的学员 | 终端体验好 | M 系列芯片偶发兼容问题 |
| **云端 Notebook** | 不想安装、设备受限 | 打开浏览器即用 | 依赖网络，文件持久化需注意 |

### 7.2 决策树

```
你的电脑能自由安装软件吗？
├── 能 → 日常用 Windows？
│         ├── 是 → 选 Windows + Anaconda 本地路径
│         └── 否 → 选 Mac + Anaconda 本地路径
└── 不能 / 不想装 → 选云端 Notebook（Colab、Kaggle 等）
```

**本课程推荐**：本地 Anaconda + Jupyter Lab。云端可作为备用，但不替代本地练习。

## 第 8 节 · Anaconda 安装与常见失败

### 8.1 安装步骤（Windows）

1. 访问 [https://www.anaconda.com/download](https://www.anaconda.com/download) 下载安装包
2. 安装时勾选 **「Add Anaconda to PATH」**（或安装后手动配置）
3. 打开 **Anaconda Prompt**（不是普通 CMD）
4. 输入 `conda --version`，看到版本号即成功

### 8.2 安装步骤（Mac）

1. 下载 macOS 版 Anaconda 或 Miniconda（更轻量，推荐）
2. 打开终端，运行安装脚本
3. 输入 `conda --version` 验证

### 8.3 常见失败与处理

| 现象 | 原因 | 处理 |
|:---|:---|:---|
| `conda` 不是内部命令 | PATH 未配置 | 重装并勾选 PATH，或手动添加 |
| 安装极慢 | 网络问题 | 换镜像源，或使用 Miniconda |
| 已有旧版 Python 冲突 | 多版本共存 | 用 conda 环境隔离，不删系统 Python |
| 权限不足 | 未用管理员/用户目录无写权限 | 换安装目录或用用户级安装 |

## 第 10 节 · 虚拟环境与编程助手部署原则

### 10.1 为什么需要虚拟环境

不同项目依赖不同版本的库。全局安装会导致「A 项目能跑，B 项目崩了」。

**原则**：一个课程 / 一个项目 = 一个独立环境。

### 10.2 创建与激活（conda）

```bash
# 创建名为 psit-alpha 的环境，Python 3.10
conda create -n psit-alpha python=3.10 -y

# 激活
conda activate psit-alpha

# 安装 Jupyter
conda install jupyterlab -y

# 注册内核（让 Jupyter 能选中这个环境）
python -m ipykernel install --user --name psit-alpha --display-name "Python (psit-alpha)"
```

### 10.3 编程助手部署原则

不推销具体产品。只讲**验收标准**：

| 该做的 | 不该做的 |
|:---|:---|
| 用助手生成草稿，自己逐行审查 | 不读代码直接复制粘贴提交 |
| 把需求写成可验收的规格 | 用「帮我做个程序」这种模糊指令 |
| 在独立环境里试运行 | 让助手直接改系统级配置 |
| 保留对话记录便于复盘 | 把 API 密钥写进 Notebook 并分享 |
